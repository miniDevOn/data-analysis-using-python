```python
file = open('ques2.txt','r+')
count=0
for word in file.read().split(" "):
    count+=1
print("The number of words are:",count)


import numpy as np
file = open('ques2.txt','r+')
wordslength=[]
for word in file.read().split(" "):
    word=word.strip(".,?:()\"")
    word=word.replace(".","")
    word=word.replace("'","")
    wordslength+=[len(word)]
wordslength.sort()
count1=np.bincount(wordslength)
print("The Mean of the words length is:",np.mean(wordslength))
print("The Median of the words length is:",np.percentile(wordslength,50))
print("The Mode of the words length is:",np.argmax(count1))
```
